```python
# Count occurrences of 9999 in Total_Work_Hour_per_Month
df["Total_Work_Hour_per_Month"].value_counts()
```

```python
# Replace 9999 with NaN
df.loc[df["Total_Work_Hour_per_Month"] == 9999, "Total_Work_Hour_per_Month"] = pd.NA
```

```python
# Convert Phone_Number to string and remove .0 artifacts
df["Phone_Number"] = df["Phone_Number"].astype(str).str.replace(r"\.0$", "", regex=True)
```

```python
# Compute Hourly Wage
df["Hourly_Wage"] = df["Salary_per_Month"] / df["Total_Work_Hour_per_Month"]

# Display results
df[["Salary_per_Month", "Total_Work_Hour_per_Month", "Hourly_Wage"]].head()
```
